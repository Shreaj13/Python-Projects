```python
#code snippet 13
# Preparing dataframe without the missing data variables -
data1 = data[['School', 'School URL', 'District', 'District URL', 'City', 'City URL',
       'Zip', 'County', 'Phone', 'Is Title I', 'Is Charter', 'Is Magnet',
       'Is Virtual', 'Number Students', 'Number Full-time Teachers',
       'Student/Teacher Ratio', 'Percent Free/Disc Lunch',
       'Percent Homes Rented', 'Percent of Unemployment',
       'Percent Parents with Masters and above education',
       'Percent parents with less than High School education',
       'Percent Home Incomes with less than Median income',
       'Standard Score']]

# Dispalying the data1 dataset columns
data1.columns
```

```python
#code snippet 15
# Creating the pridictors variables after qualitative and auantitavive data prep activities
predictors =  data1[[ 'District', 'City', 'City URL',
       'Zip', 'County','Is Title I', 'Is Charter', 'Is Magnet',
       'Is Virtual', 'Number Students', 'Number Full-time Teachers',
       'Student/Teacher Ratio', 'Percent Free/Disc Lunch',
       'Percent Homes Rented', 'Percent of Unemployment',
       'Percent Parents with Masters and above education',
       'Percent parents with less than High School education',
       'Percent Home Incomes with less than Median income',
       'Standard Score']]

predictors.columns
```

```python
# Code Snippet 16
# Installing the sweetviz EDA packeage
!pip install sweetviz
```

```python
# Code Snippet 17
# Import the sweetviz EDA packeage
import sweetviz as sv
```

```python
# Code snippet 18
# Generating EDA report
sv_report = sv.analyze(predictors)

# puting SV report in SV report in a interactive HTML format
sv_report.show_html("Texas Schools EDA report.html")
```
